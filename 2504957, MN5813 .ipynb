```python
# Here, I assign points awarded based on match results
df_json7['HomePoints'] = df_json7['FTR'].map({'H': 3, 'D': 1, 'A': 0})  # I am mapping how many points are awarded for each type of result for the Home or Away team points depending on the result. - (Mckinney, 2023, chapter 7.2)
df_json7['AwayPoints'] = df_json7['FTR'].map({'H': 0, 'D': 1, 'A': 3})  # Away team points - (Mckinney, 2023, chapter 7.2)

# I sum the stats accumulated for every home and away team in each game and assing the stats respectively to each team considering the season the match was played.
df_home_stats1 = df_json7.groupby(['Season', 'HomeTeam']).agg({  # (Reppel, 2017, Ch. 17.2.3.4)
    'HomePoints': 'sum',
    'FTHG': 'sum',  # Goals scored by the home team
    'FTAG': 'sum',  # Goals conceded against the home team
    'HS': 'sum',    # Shots for the Home team
    'HST': 'sum',   # HomeShotsOnTarge
    'HC': 'sum',    # HomeCorners
    'HF': 'sum',    # HomeFouls
    'AS': 'sum',    # AwayShots (against Home)
    'AST': 'sum',   # AwayShotsOnTarget
    'AC': 'sum',    # AwayCorners
    'AF': 'sum'     # AwayFouls (against Home)
}).reset_index()

df_away_stats1 = df_json7.groupby(['Season', 'AwayTeam']).agg({   # (Reppel, 2017, Ch. 17.2.3.4)
    'AwayPoints': 'sum',
    'FTAG': 'sum',  # Goals scored for the away team.
    'FTHG': 'sum',  # Goals conceded by the away team
    'AS': 'sum',    # AwayShots
    'AST': 'sum',   # AwayShotsOnTarget
    'AC': 'sum',    # AwayCorners
    'AF': 'sum',    # AwayFouls
    'HC': 'sum',    # Corners conceded against the away team (HomeCorners)
    'HS': 'sum',    # Shots conceded against the away team (HomeShots)
    'HST': 'sum',   # ShotsOnTarget conceded against the away team (HomeShotsOnTarget)
    'HF': 'sum'     # Fouls committed by home team 
}).reset_index()

# I rename columns to those that are more easily understandable.
df_home_stats1.rename(columns={
    'HomeTeam': 'Team',
    'FTHG': 'Goals_Scored',
    'FTAG': 'Goals_Conceded',
    'HS': 'Shots_For',
    'HST': 'Shots_OT_For',
    'HC': 'Corners_For',
    'HF': 'Fouls_Against',
    'AF': 'Fouls_For',
    'AS': 'Shots_Against',
    'AST': 'Shots_OT_Against',
    'AC': 'Corners_Against'
}, inplace=True)

df_away_stats1.rename(columns={
    'AwayTeam': 'Team',
    'FTAG': 'Goals_Scored',
    'FTHG': 'Goals_Conceded',
    'AS': 'Shots_For',
    'AST': 'Shots_OT_For',
    'AC': 'Corners_For',
    'AF': 'Fouls_Against',
    'HF': 'Fouls_For',
    'HS': 'Shots_Against',
    'HST': 'Shots_OT_Against',
    'HC': 'Corners_Against'
}, inplace=True)

# I am merging the two home and away stats data frames I created above using an inner join to find the the common set found in both tables. As both tables are identical in terms of columns combined with the calculations above, the inner join method is perfect for this example.  
df_team_stats7 = pd.merge(df_home_stats1, df_away_stats1, on=['Season', 'Team'], how='inner')      # (Mckinney, 2023, CH 8.2)

# Display final stats
print(df_team_stats7.head()) #As you can see we have two columns each for goals scored, goals conceded and so on, this is because we have calculated the home and the away stats separately to ensure accuracy and consistency in our methods. 
```

```python


# Here I am now using the home and away calculations and combining them to give for an example an overall points tally for each team = Home points for that team + Away points for that team. The values are all in the same row for each team and season as they have been grouped by season and team in the previous calculation.
# This is a simple calculation of adding the selected existing data frame column values to create a new column labelled prior to the equals sign below.

df_team_stats7['Total_Points'] = df_team_stats7['HomePoints'] + df_team_stats7['AwayPoints']
df_team_stats7['Goals_Scored'] = df_team_stats7['Goals_Scored_x'] + df_team_stats7['Goals_Scored_y']
df_team_stats7['Goals_Conceded'] = df_team_stats7['Goals_Conceded_x'] + df_team_stats7['Goals_Conceded_y']
df_team_stats7['Goal_Difference'] = df_team_stats7['Goals_Scored'] - df_team_stats7['Goals_Conceded']
df_team_stats7['Shots_For'] = df_team_stats7['Shots_For_x'] + df_team_stats7['Shots_For_y']
df_team_stats7['Shots_Against'] = df_team_stats7['Shots_Against_x'] + df_team_stats7['Shots_Against_y']
df_team_stats7['Shots_OT_For'] = df_team_stats7['Shots_OT_For_x'] + df_team_stats7['Shots_OT_For_y']
df_team_stats7['Shots_OT_Against'] = df_team_stats7['Shots_OT_Against_x'] + df_team_stats7['Shots_OT_Against_y']
df_team_stats7['Corners_For'] = df_team_stats7['Corners_For_x'] + df_team_stats7['Corners_For_y']
df_team_stats7['Corners_Against'] = df_team_stats7['Corners_Against_x'] + df_team_stats7['Corners_Against_y']
df_team_stats7['Fouls_For'] = df_team_stats7['Fouls_For_x'] + df_team_stats7['Fouls_For_y']
df_team_stats7['Fouls_Against'] = df_team_stats7['Fouls_Against_x'] + df_team_stats7['Fouls_Against_y']

# I create a new dataframe keeping the calculated columns above for analysis and the grouped by columns Team and Season. 
df_team_stats8 = df_team_stats7[['Season', 'Team', 'Total_Points', 'Goals_Scored', 'Goals_Conceded', 
                               'Goal_Difference', 'Shots_For', 'Shots_Against', 'Shots_OT_For', 
                               'Shots_OT_Against', 'Corners_For', 'Corners_Against', 
                               'Fouls_For', 'Fouls_Against']]

# Here I print the first few rows of the final aggregated data frame which I will now use to analyse the aim of What makes a champion.
print(df_team_stats8.head())
```

```python
# Below I sort the data frame first by each season in ascending order and then by total points in descending order. The ascending = True is in relation to Season (Ascending order) and the false is relevant for sorting the final table in ascending order by total points.   
df_points_stats8 = df_team_stats8.sort_values(by=['Season', 'Total_Points'], ascending=[True, False]) # (Mckinney, 2023, CH. 10.3)

# The below code has extracted the two teams in two sperate tables with the most points from each season, the first code for first place in each season and the second code for second place in each season. 
df_champions = df_points_stats8.groupby('Season').nth(0).reset_index()  # (Mckinney, 2023, CH 10.2, Table 10.1)
df_runners_up = df_points_stats8.groupby('Season').nth(1).reset_index()  # (Mckinney, 2023, CH 10.2, Table 10.1)

# I merge the champions and runners-up tables
df_champions_vs_runnersup = pd.merge(
    df_champions, df_runners_up, on='Season', suffixes=('_Champion', '_RunnerUp')  #(Mckinney, 2023, CH 8.2)
)

# I select the relevant columns to display
columns_to_display = [
    'Season', 
    'Team_Champion', 'Team_RunnerUp', 
    'Total_Points_Champion', 'Total_Points_RunnerUp', 
    'Goals_Scored_Champion', 'Goals_Scored_RunnerUp', 
    'Goals_Conceded_Champion', 'Goals_Conceded_RunnerUp', 
    'Shots_For_Champion', 'Shots_For_RunnerUp', 
    'Shots_Against_Champion', 'Shots_Against_RunnerUp',  
    'Shots_OT_For_Champion', 'Shots_OT_For_RunnerUp', 
    'Shots_OT_Against_Champion', 'Shots_OT_Against_RunnerUp', 
    'Corners_For_Champion', 'Corners_For_RunnerUp', 
    'Corners_Against_Champion', 'Corners_Against_RunnerUp', 
    'Fouls_For_Champion', 'Fouls_For_RunnerUp', 
    'Fouls_Against_Champion', 'Fouls_Against_RunnerUp'
]

# I print the new merged DataFrame with the selected columns
print(df_champions_vs_runnersup[columns_to_display])
```

```python
# Below I calculate the statistical differences between the champions and the runners-up by creating new columns in the existing data frame which are based on the subtraction calculation between the Champions and Runner's up values. 
df_champions_vs_runnersup['Points_Difference'] = (
    df_champions_vs_runnersup['Total_Points_Champion'] - df_champions_vs_runnersup['Total_Points_RunnerUp']
)
df_champions_vs_runnersup['Goal_Difference'] = (
    df_champions_vs_runnersup['Goal_Difference_Champion'] - df_champions_vs_runnersup['Goal_Difference_RunnerUp']
)
df_champions_vs_runnersup['Goals_Scored_Difference'] = (
    df_champions_vs_runnersup['Goals_Scored_Champion'] - df_champions_vs_runnersup['Goals_Scored_RunnerUp']
)
df_champions_vs_runnersup['Goals_Conceded_Difference'] = (
    df_champions_vs_runnersup['Goals_Conceded_Champion'] - df_champions_vs_runnersup['Goals_Conceded_RunnerUp']
)
df_champions_vs_runnersup['Shots_For_Difference'] = (
    df_champions_vs_runnersup['Shots_For_Champion'] - df_champions_vs_runnersup['Shots_For_RunnerUp']
)
df_champions_vs_runnersup['Shots_Against_Difference'] = (
    df_champions_vs_runnersup['Shots_Against_Champion'] - df_champions_vs_runnersup['Shots_Against_RunnerUp']
)
df_champions_vs_runnersup['Shots_OT_For_Difference'] = (
    df_champions_vs_runnersup['Shots_OT_For_Champion'] - df_champions_vs_runnersup['Shots_OT_For_RunnerUp']
)
df_champions_vs_runnersup['Shots_OT_Against_Difference'] = (
    df_champions_vs_runnersup['Shots_OT_Against_Champion'] - df_champions_vs_runnersup['Shots_OT_Against_RunnerUp']
)
df_champions_vs_runnersup['Corners_For_Difference'] = (
    df_champions_vs_runnersup['Corners_For_Champion'] - df_champions_vs_runnersup['Corners_For_RunnerUp']
)
df_champions_vs_runnersup['Corners_Against_Difference'] = (
    df_champions_vs_runnersup['Corners_Against_Champion'] - df_champions_vs_runnersup['Corners_Against_RunnerUp']
)
df_champions_vs_runnersup['Fouls_For_Difference'] = (
    df_champions_vs_runnersup['Fouls_For_Champion'] - df_champions_vs_runnersup['Fouls_For_RunnerUp']
)
df_champions_vs_runnersup['Fouls_Against_Difference'] = (
    df_champions_vs_runnersup['Fouls_Against_Champion'] - df_champions_vs_runnersup['Fouls_Against_RunnerUp']
)

# I select the new difference columns to display
columns_to_display = [
    'Season', 'Team_Champion', 'Team_RunnerUp', 
    'Points_Difference', 'Goals_Scored_Difference', 'Goals_Conceded_Difference', 
    'Goal_Difference', 'Shots_For_Difference', 'Shots_Against_Difference', 
    'Shots_OT_For_Difference', 'Shots_OT_Against_Difference', 
    'Corners_For_Difference', 'Corners_Against_Difference', 
    'Fouls_For_Difference', 'Fouls_Against_Difference'
]

# I show the comparison table with the selected new columns
print(df_champions_vs_runnersup[columns_to_display])
```

```python
import pandas as pd

# I Create a new DataFrame for Goal Conversion Rate
df_goal_conversion_table = df_team_stats8[['Season', 'Team', 'Goals_Scored', 'Shots_For']].copy()

# I Calculate the Goal Conversion Rate
df_goal_conversion_table['Goal_Conversion_Rate'] = df_goal_conversion_table['Goals_Scored'] / df_goal_conversion_table['Shots_For']

# I count the number of seasons each team has participated in as some teams get relegated over seasons and also come back through promotion from the second tier of English football. 
df_seasons_per_team = df_goal_conversion_table.groupby('Team')['Season'].nunique().reset_index()
df_seasons_per_team.columns = ['Team', 'Seasons_Played']

# I add up all the Goal Conversion Rates and merge it with the count of seasons
df_goal_conversion_summary = df_goal_conversion_table.groupby('Team')['Goal_Conversion_Rate'].sum().reset_index()
df_goal_conversion_summary = pd.merge(df_goal_conversion_summary, df_seasons_per_team, on='Team')

# I calculate the average Goal Conversion Rate per season
df_goal_conversion_summary['Avg_Goal_Conversion_Per_Season'] = (
    df_goal_conversion_summary['Goal_Conversion_Rate'] / df_goal_conversion_summary['Seasons_Played']
)

# I created a new sorted DataFrame by average Goal Conversion Rate
df_goal_conversion_summary_sorted = df_goal_conversion_summary.sort_values('Avg_Goal_Conversion_Per_Season', ascending=False)

# I display the table
print(df_goal_conversion_summary_sorted)
```

```python
# I create a new DataFrame for referee statistics using the cleaned DataFrame I created called df_json5. 
df_referee_stats = df_json5.groupby('Referee').agg({
    'HF': 'sum',  # Home Fouls
    'AF': 'sum',  # Away Fouls
    'HY': 'sum',  # Home Yellow Cards
    'AY': 'sum',  # Away Yellow Cards
    'HR': 'sum',  # Home Red Cards
    'AR': 'sum',  # Away Red Cards
}).reset_index()

# I add the accurate match count for each match refereed
df_referee_stats['Matches_Refereed'] = df_json5.groupby('Referee').size().values

# I rename columns for clarity
df_referee_stats.rename(columns={
    'HF': 'Home_Fouls',
    'AF': 'Away_Fouls',
    'HY': 'Home_Yellow_Cards',
    'AY': 'Away_Yellow_Cards',
    'HR': 'Home_Red_Cards',
    'AR': 'Away_Red_Cards'
}, inplace=True)

# I calculate the total fouls, yellow cards, and red cards 
df_referee_stats['Total_Fouls'] = df_referee_stats['Home_Fouls'] + df_referee_stats['Away_Fouls']
df_referee_stats['Total_Yellow_Cards'] = df_referee_stats['Home_Yellow_Cards'] + df_referee_stats['Away_Yellow_Cards']
df_referee_stats['Total_Red_Cards'] = df_referee_stats['Home_Red_Cards'] + df_referee_stats['Away_Red_Cards']

# Display the new DataFrame
print(df_referee_stats)
```

```python
import matplotlib.pyplot as plt
import numpy as np

# I select the data for the bar chart
categories = ['Yellow Cards', 'Yellow Cards', 'Red Cards', 'Red Cards']
teams = ['Home', 'Away', 'Home', 'Away']
values = [
    df_referee_stats['Home_Yellow_Cards'].sum(),
    df_referee_stats['Away_Yellow_Cards'].sum(),
    df_referee_stats['Home_Red_Cards'].sum(),
    df_referee_stats['Away_Red_Cards'].sum(),     #(Mckinney, 2023, CH 9.1)
]

# I add distinct colours for the bars
colors = ['#FFD700', '#FFCC00', '#FF9999', '#FF0000']  # These are the colour codes for light yellow, dark Yellow, light red, dark red

# I size and plot the chart
fig, ax = plt.subplots(figsize=(10, 6))

# I arrange the bar chart using the categories I created above
x = np.arange(len(categories))

# I create the bar plot
bars = ax.bar(x, values, color=colors, edgecolor='black')

# I add annotations for the values on the bars
for bar in bars:
    height = bar.get_height()
    ax.annotate(f'{int(height)}',
                xy=(bar.get_x() + bar.get_width() / 2, height),
                xytext=(0, 3),  # 3 points vertical offset
                textcoords="offset points",
                ha='center', fontsize=10)

# I design the plot
ax.set_title('Comparison of Cards Given to Home vs. Away Teams', fontsize=16)
ax.set_xlabel('Card Type', fontsize=14)
ax.set_ylabel('Total Count', fontsize=14)
ax.set_xticks(x)
ax.set_xticklabels([f"{team} {category}" for team, category in zip(teams, categories)], fontsize=12)
ax.tick_params(axis='y', labelsize=12)
plt.tight_layout()

# I display the chart
plt.show()
```

**Conclusion**

This study identified goals scored and goals conceded as the most consistent key performance indicators in differentiate the champions from the runners-up in the Premier League during the past decade. Metrics for and against teams such as shots, fouls and corners exhibited substantial variability, suggesting a restricted reliability as performance forecasters. Notable anomalies like the 2017/8 season, where the champions, Manchester City, dominated across the metrics and the 2019/20 season, where, runners-up, the same team led in all the metrics but the points tally tell the tale of the complexity and dynamism of football team performance. 

The findings of the study when interpreting goal conversion rates showed key insights in attacking efficiency, 8-10% was around the average team performance with most efficient teams attaining at least 12%. “Fine margins in football” is. A term you regularly hear in football which is true here especially in the scoring statistical analysis department of football.
The referee performance analysis highlights some potential biases or possibly influence from the home crowd as away teams consistently received more cads despite similar foul counts to home teams in the last decade. Although there was an argument to advocate for uniformed decision-making as the data visualisations supported the consistency in cards issued per match, 3-4 cards. The disparity between the cards issued to the away teams compared to the home teams does raise questions about the contextual influences of a football match on referee behaviour and decision-making. 

These findings provide a foundation for further, more comprehensive analyses. Predictive modelling using advanced machine learning techniques could forecast key outcomes like goal conversion rates for players as well as teams. Also further analysis of contextual factors like weather, time and fixture congestion could better develop a better understanding of the interactions of contextual factors with key performance metrics.
